## Correlation and Collinearity Removal

```python
cols=['U_OB_OFNT_VC_DUR_T6','U_OB_ONT_VC_DUR_T6','U_OB_SMS_PAID_CNT_T6','U_OB_SMS_FREE_CNT_T6','U_GPRS_VOL_T6','RECH_AMT_T6']
data=pd.DataFrame(data=df,columns=cols)
data.columns=['OFFnet_Duration','ONnet_Duration','Paid Sms','Free Sms','Gprs','Recharge']

data.head()
```

```python
# Create correlation matrix
corr_matrix = data.corr().abs()
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
# # Drop features 
corred=data.drop(data[to_drop], axis=1)
corred.dropna(inplace= True)
#corred['Revenue']=corred[['U_OB_ONT_SMS_REV_T6','U_OB_OFNT_SMS_REV_T6','U_GPRS_REV_T6','U_OB_VC_REV_T6']].sum(axis=1)
corred.head()
```

## Scalling and then Clustering 

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans #imports Kmean algorithm from the sklearn's cluster library
from sklearn import metrics
scaler = MinMaxScaler()
```

```python
temp=scaler.fit_transform(corred)
scaled=pd.DataFrame(columns=corred.columns,data=temp)
scaled.head()
```

```python
pred=scaled.copy()
kmean = KMeans(n_clusters=4, random_state=0)
kmean.fit(scaled)
pred['kmean'] = kmean.labels_
print(pred.shape)
print(pred['kmean'].value_counts())
```

```python
from sklearn import metrics
metrics.silhouette_score(scaled, kmean.labels_)
```

```python
def sse(x, means):
    return np.sum((x.values - means[x['pred']])**2)
n = []
ss = []
for i in range(30):
    d = scaled.copy()
    k = KMeans(n_clusters=i+1, random_state=0).fit(d)
    d['pred'] = k.labels_
    means = dict()
    for c in d['pred'].unique(): # cluster mean ko uthao= aik value 
        m = d[d['pred'] == c].mean().values
        means[c] = m
    d['sse'] = d.apply(lambda x: sse(x, means), axis=1)
    n.append(i+1)
    ss.append(d['sse'].sum()) # as formula errors are summed
plt.figure(figsize=(15,10));
plt.plot(n, ss);
plt.xlabel('Value of K');
plt.ylabel('Error');
plt.title('Elbow Analysis for optimal value of K');
```
